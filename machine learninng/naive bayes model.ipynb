```python
import re
import pandas as pd
```

```python
with open(r"C:\Users\USER\Downloads\university\assignment\year 2\Machine learning\MachineLearningCW\machine learninng\adult.names") as file:
    text = file.read()
    labels = re.findall(r'\n(\w*_?\W?):', text)

Data_set = pd.read_csv(
    r"C:\Users\USER\Downloads\university\assignment\year 2\Machine learning\MachineLearningCW\machine learninng\adult.data",
    header=None,
    names=labels + ['income']
)

Data_array = Data_set.values
# print(Data_array)
Data_set.head()
```

```python
print("No of rows in dataset before preprocessing : ", len(Data_set))
```

### Find duplicates

```python
Data_set.duplicated().values.any()
```

```python
print("Duplicate Rows except first occurrence:")

duplicates = Data_set[Data_set.duplicated()]
duplicates
```

### Drop duplicates

```python
# Drop duplicates
Data_set_no_duplicates = Data_set.drop_duplicates()
Data_set_no_duplicates
```

```python
print("No of rows in dataset after removing duplicates : ", len(Data_set_no_duplicates))
```

### Check for missing values

```python
# Check for missing values in the dataset
missing_values = Data_set_no_duplicates.isnull().sum()
print("\nMissing Values in the Dataset:")
print(missing_values)
```

### Handle categorical variables

```python
categorical_columns = Data_set_no_duplicates.select_dtypes(include=['object']).columns
Data_set_no_duplicates.loc[:, categorical_columns] = Data_set_no_duplicates.loc[:, categorical_columns].astype('category')
Data_set_no_duplicates = pd.get_dummies(Data_set_no_duplicates, drop_first=True)
```

### Label Encoding for the Target Variable:

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
Data_set_no_duplicates['income_ >50K'] = label_encoder.fit_transform(Data_set_no_duplicates['income_ >50K'])
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
numerical_columns = Data_set_no_duplicates.select_dtypes(include=['int64', 'float64']).columns
Data_set_no_duplicates[numerical_columns] = scaler.fit_transform(Data_set_no_duplicates[numerical_columns])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot histograms for numerical features
numerical_features = Data_set_no_duplicates.select_dtypes(include=['int64', 'float64']).columns
Data_set_no_duplicates[numerical_features].hist(bins=20, figsize=(15, 10))
plt.suptitle('Histograms of Numerical Features', y=1.02)
plt.show()
```
